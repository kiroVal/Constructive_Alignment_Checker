```python
import os
from docx import Document
import pandas as pd
import io

# In a standard Python environment, you would get the file path differently.
# For example, you can prompt the user for the file path.
file_path = input("Please enter the path to your DOCX file: ")

# Load the DOCX document
document = None  # Initialize document to None
try:
    document = Document(file_path)
    print(f"Successfully loaded '{file_path}'")
except Exception as e:
    print(f"Error loading document: {e}")

# Proceed only if the document was loaded successfully
if document is not None:
    # Define keywords to identify relevant tables
    keywords = {
        "LEARNING OUTCOMES": ["learning outcomes"],
        "DELIVERABLES/OUTCOMES": ["deliverables", "outcomes"],
        "ASSESSMENT": ["assessment"]
    }

    target_tables_data = {}

    for table in document.tables:
        for category, keyword_list in keywords.items():
            if table.rows:
                first_row_text = "".join([cell.text.strip().lower() for cell in table.row_cells(0)])
                if any(keyword in first_row_text for keyword in keyword_list):
                    table_data = []
                    for row in table.rows:
                        row_data = [cell.text.strip() for cell in row.cells]
                        table_data.append(row_data)

                    if category in target_tables_data:
                        target_tables_data[category].extend(table_data)
                    else:
                        target_tables_data[category] = table_data

                    print(f"Found table for: '{category}'")

    all_data = []
    for category, data in target_tables_data.items():
        df = pd.DataFrame(data)
        df['Source Table'] = category
        all_data.append(df)

    if all_data:
        combined_df = pd.concat(all_data, ignore_index=True)
        print("\nCombined Data (first 5 rows):")
        print(combined_df.head()) # Use print instead of display in standard Python environment

        csv_filename = "extracted_course_outline_tables.csv"
        try:
            # Create an in-memory text buffer
            csv_buffer = io.StringIO()

            # Save the combined DataFrame to the buffer in CSV format
            combined_df.to_csv(csv_buffer, index=False)

            # Get the CSV content as a string
            csv_content = csv_buffer.getvalue()

            # Write the CSV content to a file in the current directory
            with open(csv_filename, 'w') as f:
                f.write(csv_content)

            print(f"\nCombined data saved to '{csv_filename}'")

            # You won't have a downloadable link like in Colab,
            # the file will be saved to the same directory as your Python script.

        except Exception as e:
            print(f"Error saving CSV file: {e}")

    else:
        print("No target tables found.")
```

```python
course_outline_start_index = -1

for i, paragraph in enumerate(document.paragraphs):
    if "COURSE OUTLINE" in paragraph.text.strip().upper():
        course_outline_start_index = i
        print(f"Found 'COURSE OUTLINE' paragraph at index {i}")
        break

if course_outline_start_index == -1:
    print("Could not find 'COURSE OUTLINE' section.")
```

```python
tables_in_course_outline = []

# Get the body of the document
document_body = document.element.body

# Find the index of the paragraph element in the body
course_outline_element_index = -1
for i, element in enumerate(document_body):
    if element == document.paragraphs[course_outline_start_index]._element:
        course_outline_element_index = i
        break

if course_outline_element_index != -1:
    # Iterate through the body elements after the COURSE OUTLINE paragraph element
    for i in range(course_outline_element_index + 1, len(document_body)):
        element = document_body[i]

        # Check if the element is a table
        if element.tag.endswith('tbl}'):
            # Find the corresponding table object
            table = None
            for doc_table in document.tables:
                if doc_table._element == element:
                    table = doc_table
                    break

            if table:
                tables_in_course_outline.append(table)
                print(f"Found a table in the COURSE OUTLINE section.")

if not tables_in_course_outline:
    print("No tables found in the COURSE OUTLINE section.")
else:
    print(f"Found {len(tables_in_course_outline)} tables in the COURSE OUTLINE section.")
```

```python
target_tables = {}
target_titles = ["LEARNING OUTCOMES", "DELIVERABLES", "ASSESSMENT"]

# Iterate through all tables in the document
for i, table in enumerate(document.tables):
    if table.rows:
        # Check if the first row contains any of the target titles as substrings
        first_row_text = "".join([cell.text.strip().upper() for cell in table.row_cells(0)])
        for title in target_titles:
            if title in first_row_text:
                # Store the table with its identified title
                target_tables[title] = table
                print(f"Found table '{title}' (found in document)")
                # Remove the found title from the list to avoid finding duplicates for the same table
                if title in target_titles:
                    target_titles.remove(title)
                break # Move to the next table once a title is found

if not target_tables:
    print("No target tables found using substring match in any table.")
else:
    print(f"Found {len(target_tables)} target tables in the document.")
```

```python
import pandas as pd

all_data = []

for title, table in target_tables.items():
    table_data = []
    for row in table.rows:
        row_data = [cell.text.strip() for cell in row.cells]
        table_data.append(row_data)

    # Convert table data to a DataFrame and add a 'Source Table' column
    df = pd.DataFrame(table_data)
    df['Source Table'] = title
    all_data.append(df)

if all_data:
    combined_df = pd.concat(all_data, ignore_index=True)
    print("\nCombined Data (first 5 rows):")
    display(combined_df.head())
else:
    print("No data to combine from target tables.")
```

```python
import io
from IPython.display import FileLink

# Create an in-memory text buffer
csv_buffer = io.StringIO()

# Save the combined DataFrame to the buffer in CSV format
combined_df.to_csv(csv_buffer, index=False)

# Get the CSV content as a string
csv_content = csv_buffer.getvalue()

# Define the filename for the download
csv_filename = "Extracted_Course_Outline_Tables.csv"

# Write the CSV content to a file in the current directory
with open(csv_filename, 'w') as f:
    f.write(csv_content)

# Create and display a downloadable link
display(FileLink(csv_filename))
```
